# Формирование модели монетизации игрового приложения

## Введение  

   Заказчик - маркетинговый отдел компании-разработчика игр для смартфонов, планирует рекламный бюджет для своей новой игры. Для этих целей продакт менеджер компании запросил аналитический отчет, на основе которого должна быть сформирована модель монетизации игрового приложения. Данная модель монетизации должна отражать обоснование выбора сегмента приложения для показа рекламы таким образом, чтобы рекламная нагрузка на пользователя была оптимальной, что подразумевает предотвращение оттока пользователей наравне с обеспечением максимально возможного показа рекламы.   
   Механика игры подразумевает колонизацию галактики путем прохождения уровней через строительство зданий и сражение с противниками. На этапе планирования модели монетизации, заказчиком предложено размещение рекламы на экране с выбором типа объекта для постройки. Для выполнения аналитического проекта по формированию оптимальной модели монетизации, заказчик предоставил данные об игровых событиях, рекламе и пользователях за период прохождения пользователями первого уровня игры.  
   
1. Базовая модель монетизации - размещение рекламы на этапе выбора объекта строительства.
2. Альтернативная модель монетизации - размещение рекламы на этапе выбора объекта строительства, а также на этапе достижения пользователем статуса завершения первого уровня.

### Цель проекта

Сформировать модель монетизации игрового приложения

2. Оценить влияние событий на совершение целевого события

### Описание данных

Датасет *game_actions.csv*:

- `event_datetime` — время события;
- `event` — одно из трёх событий:
    1. `building` — объект построен,
    2. `finished_stage_1` — первый уровень завершён,
    3. `project` — проект завершён;
- `building_type` — один из трёх типов здания:
    1. `assembly_shop` — сборочный цех,
    2. `spaceport` — космопорт,
    3. `research_center` — исследовательский центр;
- `user_id` — идентификатор пользователя;
- `project_type` — тип реализованного проекта;

Датасет *ad_cost.csv*:

- `day` - день, в который был совершен клик по объявлению
- `source` - источник трафика
- `cost` - стоимость кликов

Датасет *user_source.csv*:

- `user_id` - идентификатор пользователя
- `source` - источников, с которого пришёл пользователь, установивший приложение

### Замена типов данных

```python
game_actions['event_datetime']=pd.to_datetime(game_actions['event_datetime'], format='%Y-%m-%d %H:%M:%S',errors='coerce')
ad_cost['day']=pd.to_datetime(ad_cost['day'],format='%Y-%m-%d',errors='coerce').dt.date

display(f'Игровые события происходили за период с {game_actions["event_datetime"].min()} по {game_actions["event_datetime"].max()}')
display(f'Данные об оплате трафика включают период с {ad_cost["day"].min()} по {ad_cost["day"].max()}')

display(f'Типы игровых событий (event) : {game_actions["event"].unique()}')
display(f'Типы зданий: {game_actions["building_type"].unique()}')
display(f'Источники трафика: {ad_cost["source"].unique()}')
```

```python
#сформируем список пользователей, которые завершили игру проектом
list_project_users = game_actions.loc[(game_actions["event"]=="project"), 'user_id'].tolist()

#сформируем список пользователей, которые не завершили проект, для того чтобы в дальнейшем отфильтровать по ним 
#пользователей, прошедших первый уровень через победу в сражении с противником
list_non_project_users = game_actions.query('user_id not in @list_project_users & event == "finished_stage_1"')['user_id'].to_list()
```

```python
game_actions['finish'] = np.nan # колонка для определения статуса прохождения первого уровня

game_actions.loc[game_actions['user_id'].isin(list_project_users),'finish'] = 'finish_project'
game_actions.loc[game_actions['user_id'].isin(list_non_project_users),'finish'] = 'finish_war'
game_actions['finish'] = game_actions['finish'].fillna('not_finish')
```

```python
#Поле building_tupe актуально только для событий building, 
# следовательно все пропуски в данной колонке мы можем заполнить "заглушкой" "no_building". 
game_actions['building_type'] = game_actions['building_type'].fillna('no_buildings')
```

```python
# project_type может быть только 1, все пропуски заполним заглушкой "no_station"
game_actions['project_type'] = game_actions['project_type'].fillna('no_station')

display(game_actions['project_type'].unique())
display(game_actions.info())
```

```python
# определим наираннюю и наипозднюю даты событий для кажого пользователя, 
# вычислим разницу между поздней и ранней датой 
# для пользователей, со статусом finish_war или finish_project 
# эта разница будет равна скорости прохождения 1го уровня
min_event = game_actions.groupby('user_id',as_index=False)['event_datetime'].min()
max_event = game_actions.groupby('user_id',as_index=False)['event_datetime'].max()

diff_event = min_event.merge(max_event, how='inner', on='user_id',suffixes=('_min', '_max'))
diff_event['game_time'] = diff_event['event_datetime_max'] - diff_event['event_datetime_min']
diff_event['game_time'] = diff_event['game_time'].dt.days
display(diff_event.head(1))

ready_set = game_actions.merge(diff_event, how='outer', on='user_id')
display(ready_set.head(1))
```

```python
# добавим к ready_set данные об источнике трафика


ready_set = ready_set.merge(user_source, how='outer', on='user_id')

display(ready_set.head(1))
```

```python
ready_set.to_csv('/home/daniil/Downloads/final_project/ready_set.csv')
```

```python
# сырая таблица профилей

profiles = ready_set#.dropna(subset=['building_type'])#исключаем тех кто не строит-т.е. кто не смотрит рекламу
#profiles = profiles[profiles['building_type']!='no_buildings']#исключаем тех кто не строит-т.е. кто не смотрит рекламу

profiles = profiles.sort_values(by=['user_id','event_datetime'])
profiles['ad_revenue'] = 0.07 # добавляем столбец с деньгами получаемыми от просмотра рекламы

profiles['lifetime'] = (
    profiles['event_datetime'] - profiles['event_datetime_min']
    ).dt.days # добавляем столбец с lifetime пользователя

profiles['min_day']=pd.to_datetime(profiles['event_datetime_min'],  #удаляем время из даты первой постройки
                                       format='%Y-%m-%d',errors='coerce').dt.date  
display(profiles.head(1))
```

### Расчет стоимости привлечения клиента (CAC).

```python
# определим функцию для расчета стоимости привлечения клиента
def cac_profiles(cac_profiles, ad_cost):
    
    cac_profiles = (
        profiles.sort_values(by=['user_id', 'event_datetime_min'])
        .groupby('user_id')
        .agg({'event_datetime_min': 'first', 'source': 'first'})
        .rename(columns={'event_datetime_min': 'event_datetime_min'})
        .reset_index()  # возвращаем user_id из индекса
    )

    # определяем дату первого посещения
    # и первый день месяца, в который это посещение произошло
    # эти данные понадобятся для когортного анализа
    cac_profiles['day'] = cac_profiles['event_datetime_min'].dt.date
    cac_profiles['day'] = cac_profiles['event_datetime_min'].astype('datetime64[D]')
    
    new_users = (
        cac_profiles.groupby(['day', 'source'])
        .agg({'user_id': 'nunique'})
         # столбец с числом пользователей назовём unique_users
        .rename(columns={'user_id': 'unique_users'})
        .reset_index())  

    cac_ad_costs = ad_cost
    cac_ad_costs['day'] += timedelta(days=1)

    cac_ad_costs['day']  = cac_ad_costs['day'].astype('datetime64[D]')

    cac_ad_costs = cac_ad_costs.merge(new_users, on=['day', 'source'], how='left')

    # делим рекламные расходы на число привлечённых пользователей
    # результаты сохраним в столбец acquisition_cost (CAC)
    cac_ad_costs['acquisition_cost'] = cac_ad_costs['cost'] / cac_ad_costs['unique_users']

    # добавим стоимость привлечения в профили
    cac_profiles = cac_profiles.merge(
        cac_ad_costs[['day', 'source', 'acquisition_cost']],
        on=['day', 'source'],
        how='left',
    )
    cac_profiles['acquisition_cost'] = cac_profiles['acquisition_cost'].fillna(0)

    return cac_profiles
```

### Расчет пожизненной ценности клиента (LTV) 

```python
#формируем профили пользователей для расчета ltv, т.е. добавляем графу со стоимостью рекламы

profiles_ltv = profiles[profiles['building_type']!='no_buildings']#исключаем тех кто не строит-т.е. кто не смотрит рекламу

display(profiles_ltv.head(1))
```

```python
# Определяем момент и горизонт анализа данных
observation_date = datetime(2020, 6, 5).date()
horizon_days = 14 # 14, т.к. это медиана количества показов рекламы для пользователей построивших станцию

# создаём опцию «игнорировать горизонт»
ignore_horizon = False
```

```python
# формируем функцию get_ltv() 
def get_ltv(
    profiles,  #профили и данные о показах рекламы
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )

    # исключаем пользователей, пришедших после максимальной даты привлечения
    result_raw = profiles.query('event_datetime_min <= @last_suitable_acquisition_date')
    
    # группируем по каналам
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort'] 
    #dimensions = dimensions + ['source']

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):

        # таблица выручки
        # строим «треугольную» таблицу
        result = df.pivot_table(
            index=dims,
            columns='lifetime',
            values='ad_revenue',  # в ячейках — выручка за каждый лайфтайм
            aggfunc='sum',
        )

        # сумма выручки с накоплением
        result = result.fillna(0).cumsum(axis=1)

        # размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )

        # объединяем размеры когорт и таблицу выручки
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)

        # считаем LTV
        # делим каждую «ячейку» в строке на размер когорты
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes
        
        return result
    
        
    result_grouped = group_by_dimensions(
        result_raw, dimensions, horizon_days
    )

    return (result_grouped, result_raw) 
```

## Проверка продуктовых гипотез.

### Разница в скорости прохождения уровня между пользователями.

План:

1. Необходимо проверить - различается ли среднее время прохождения уровня между пользователями, которые заканчивают уровень через реализацию проекта, и пользователями, которые заканчивают уровень победой над другим игроком. Для этого установим по умолчанию утверждение о том, что  время прохождения уровня одинаково в обеих группах.    
2. Следовательно нулевая гипотеза(𝐻0): время прохождения уровня пользователями, которые заканчивают уровень через реализацию проекта, равно времени прохождения уровня пользователями, которые заканчивают уровень победой над другим игроком.  
3. Альтернативная гипотеза (𝐻1) : время прохождения уровня пользователями, которые заканчивают уровень через реализацию проекта, НЕ равно времени прохождения уровня пользователями, которые заканчивают уровень победой над другим игроком. (Гипотеза двусторонняя).  
4. Критерий для проверки нулевой гипотезы : критерий Стьюдента для несвязанных выборок.  
5. Определить наблюдаемое значение статистики.  
6. Посчитать p-value. Сравнить p-value с фиксированным уровнем значимости (0.05).  
7. Вывод

```python
display(ready_set.head(1))
group_project = ready_set.query('finish == "finish_project"')
group_project = group_project.groupby('user_id')['game_time'].first()

group_war = ready_set.query('finish == "finish_war"')
group_war = group_war.groupby('user_id')['game_time'].first()
```

```python
alpha = 0.05

task_1 = st.ttest_ind(
    group_project,
    group_war)

display(f'Кол-во пользователей завершивших проект :{len(group_project)}, медиана количества дней прохождения 1 уровня :{group_project.mean()}')
display(f'Кол-во пользователей победивших противника :{len(group_war)}, медиана количества дней прохождения 1 уровня :{group_war.mean()}')

display('p-значение:', task_1.pvalue)

if task_1.pvalue< alpha:
    display("Отвергаем нулевую гипотезу")
else:
    display("Не удалось отвергнуть нулевую гипотезу") 
```

### Разница в количестве построенных объектов между пользователями.

План:

1. Необходимо проверить - различается ли среднее количество построенных объектов между пользователями, пришедшими по трафику instagram_new_adverts и youtube_channel_reklama. Для этого установим по умолчанию утверждение о том, что количество построенных объектов одинаково в обеих группах.      
2. Следовательно нулевая гипотеза (𝐻0): количество построенных объектов в группе пользователей, пришедших по трафику instagram_new_adverts равно количеству построенных объектов пользователями, пришедших по трафику youtube_channel_reklama.    
3. Альтернативная гипотеза (𝐻1): количество построенных объектов в группе пользователей, пришедших по трафику instagram_new_adverts НЕ равно количеству построенных объектов пользователями, пришедших по трафику youtube_channel_reklama. (Гипотеза двусторонняя).    
4. Критерий для проверки нулевой гипотезы : критерий Стьюдента для несвязанных выборок.    
5. Определить наблюдаемое значение статистики.    
6. Посчитать p-value. Сравнить p-value с фиксированным уровнем значимости (0.05).    
7. Вывод  

```python
group_inst = ready_set.query('source == "instagram_new_adverts" and event == "building"')
group_inst = group_inst.groupby('user_id')['event'].count()

group_youtube = ready_set.query('source == "youtube_channel_reklama" and event == "building"')
group_youtube = group_youtube.groupby('user_id')['event'].count()

display(f'Кол-во пользователей от instagram:{len(group_inst)}, медиана кол-ва построенных зданий:{group_inst.mean()}')
display(f'Кол-во пользователей от youtube:{len(group_youtube)}, медиана кол-ва построенных зданий:{group_youtube.mean()}')


```

```python
fig, ax = plt.subplots(figsize=(12,6))
columns = [group_inst, group_youtube]
ax.boxplot(columns, notch=True, patch_artist=True)
ax.set(ylabel = "Количество зданий"),
plt.xticks([1, 2], ["instagram", "youtube"])
plt.show()
```

```python
task_2 = st.ttest_ind(
    group_inst,
    group_youtube)

display('p-значение:', task_2.pvalue)

if task_2.pvalue < alpha:
    display("Отвергаем нулевую гипотезу")
else:
    display("Не удалось отвергнуть нулевую гипотезу") 
```

***Вывод***  
   
   На имеющихся данных, на уровне значимости 5% (уровне доверия 95%) есть основания отвергнуть нулевую гипотезу в пользу альтернативной.
Наблюдается тенденция к большему количеству построенных зданий пользователями, привлеченными по трафику instagram, чем пользователями, привлеченными по трафик youtube.
