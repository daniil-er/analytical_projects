### –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∞–Ω–æ–º–∞–ª–∏–π

```python
# –∑–∞ –≤—ã–±—Ä–æ—Å—ã –∏–ª–∏ –∞–Ω–æ–º–∞–ª–∏–∏ –ø—Ä–∏–º–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤, –≤—ã—Ö–æ–¥—è—â–∏–µ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã ¬±1,5 –º–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω–æ–≥–æ —Ä–∞–∑–º–∞—Ö–∞

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏ –¥–∞—Ç–∞—Å–µ—Ç–∞ –ø–æ —Ü–µ–ª–µ–≤—ã–º –∑–Ω–∞—á–µ–Ω–∏—è–º, –ª–µ–∂–∞—â–∏–º –Ω–∏–∂–µ –∏–ª–∏ –≤—ã—à–µ –≥—Ä–∞–Ω–∏—Ü—ã 1,5 –º–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω–æ–≥–æ —Ä–∞–∑–º–∞—Ö–∞

def out_icr(df,column):
    Q1 = df[column].quantile(0.25) # –ø–µ—Ä–≤—ã–π –∫–≤–∞—Ä—Ç–∏–ª—å
    Q3 = df[column].quantile(0.75) # 3–π –∫–≤–∞—Ä—Ç–∏–ª—å
    IQR = Q3 - Q1
    ser = df[column]
    
    # –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –¥–∞—Ç–∞—Å–µ—Ç–∞ –ø–æ –∫—Ä–∏—Ç–µ—Ä–∏—é : "–ø–∞—Ä–∞–º–µ—Ç—Ä < Q1-1.5IQR"
    
    filtered = df.query('@ser <= (@Q1 - 1.5 * @IQR) | @ser >= (@Q3 + 1.5 * @IQR)')
    return filtered
```

3. max_depth: –≠—Ç–æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä –æ–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ—Ç –≥–ª—É–±–∏–Ω—É –¥–µ—Ä–µ–≤–∞, —Ç–æ –µ—Å—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–∑–ª–æ–≤, —á–µ—Ä–µ–∑ –∫–æ—Ç–æ—Ä–æ–µ –º–æ–∂–µ—Ç –ø—Ä–æ–π—Ç–∏ –¥–µ—Ä–µ–≤–æ. –û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –≥–ª—É–±–∏–Ω—ã –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å —É–ø—Ä–∞–≤–ª—è—Ç—å —Å–ª–æ–∂–Ω–æ—Å—Ç—å—é –º–æ–¥–µ–ª–∏ –∏ —É–º–µ–Ω—å—à–∏—Ç—å –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ.

```python
best_forest_model = None
best_forest_result = 0
best_est = 0
best_depth = 0
for feature in list(range(1,features_train.shape[1])):
    for est in range(10, 53, 10):
        for depth in range (1, 11):
            forest_model = RandomForestClassifier(criterion='entropy',
                                                  random_state=12345,
                                                  max_features = feature,
                                                  n_estimators=est, 
                                                  max_depth=depth)
            forest_model.fit(features_train,target_train) # –æ–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
            forest_predictions = forest_model.predict(features_valid) # –ø–æ–ª—É—á–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–∏
            forest_result = accuracy_score(forest_predictions,target_valid)# —Å—á–∏—Ç–∞–µ–º –∫–∞—á–µ—Å—Ç–≤–æ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–∏
        
            if forest_result > best_forest_result:
                best_forest_model = forest_model
                best_forest_result = forest_result
                best_est = est
                best_depth = depth

display("Accuracy –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞:", best_forest_result)
```

```python
# —Å–æ–∑–¥–∞—é —Å–µ—Ç–∫—É –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤, –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –∫–æ—Ç–æ—Ä—ã—Ö —Ö–æ—á—É –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º –ø—Ä–æ–≤–µ—Ä–∏—Ç—å

param_grid_lr = { 
    'max_iter': [10, 50, 100, 200], 
    'penalty' : ['l1', 'l2'],
    'C' : np.logspace(-4, 4, 20),
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],   
    'class_weight': ['balanced']                                    
}

# —Å–æ–∑–¥–∞—é –æ–±—ä–µ–∫—Ç-–º–æ–¥–µ–ª—å, –∫–æ—Ç–æ—Ä–∞—è –≤–∫–ª—é—á–∞–µ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥—É—Ç –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞—Ç—å—Å—è —Å–∞–º–∏ —Å–æ–±–æ–π
# –∏ –≤—ã–¥–∞–¥—É—Ç –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç

logModel_grid = GridSearchCV(estimator=LogisticRegression(random_state=1234), 
                             param_grid=param_grid_lr, 
                             verbose=True, 
                             cv=5, 
                             n_jobs=-1)

logModel_grid.fit(features_train,target_train) # –æ–±—É—á–∞—é –º–æ–¥–µ–ª—å

logModel_grid_predictions = logModel_grid.predict(features_valid) # –¥–µ–ª–∞—é —Ä–∞—Å—á–µ—Ç/–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ

display(logModel_grid.best_estimator_)

logModel_grid_result = accuracy_score(logModel_grid_predictions,target_valid) # —Å—á–∏—Ç–∞—é accuracy
display(logModel_grid_result)
```

#### –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –Ω–∞ –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö

```python
# —Å–æ–∑–¥–∞–¥–∏–º –æ–±—ä–µ–∫—Ç scaler –∏ –æ–±—É—á–∏–º StandartScaler –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
scaler = StandardScaler()
scaler.fit(features_train)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –æ–±—É—á–∞—é—â–∏–π –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã–µ –Ω–∞–±–æ—Ä—ã –¥–∞–Ω–Ω—ã—Ö
feat_train_scale = scaler.transform(features_train)
feat_valid_scale = scaler.transform(features_valid)
feat_test_scale = scaler.transform(features_test)
```

```python
# –∑–¥–µ—Å—å –∫–æ–¥ —Ç–æ—á–Ω–æ —Ç–∞–∫–æ–π –∂–µ –∫–∞–∫ –≤ –ø—É–Ω–∫—Ç–µ 4.3.1., 
# –∑–∞–º–µ–Ω–∏–ª —Ç–æ–ª—å–∫–æ –æ–±—É—á–∞—é—â—É—é, –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ 
# –Ω–∞–∑–≤–∞–Ω–∏—è –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö, —á—Ç–æ–±—ã –Ω–µ —Å–ª—É—á–∞–π–Ω–æ –Ω–µ –ø–µ—Ä–µ—Å–µ–∫–∞–ª–∏—Å—å —Å –ø—Ä–µ–¥—ã–¥—É—â–∏–º–∏
# –≤—ã–±–æ—Ä–∫–∏ —Å —Ü–µ–ª–µ–≤—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ –Ω–µ –º–µ–Ω—è–ª

best_log_model_scale = None
best_log_scale_result = 0
best_depth = 0
for iteration in range(20,100,20):
    for pen in ['l2', 'l1']:
        for solv in solvers:
            try: 
                log_model_scale = LogisticRegression(random_state=12345, 
                                           max_iter=iteration,
                                           penalty=pen,
                                           solver=solv) 
                log_model_scale.fit(feat_train_scale,target_train) # –æ–±—É—á–∞–µ–º –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
                log_scale_predictions = log_model_scale.predict(feat_valid_scale) # –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–∏
                log_scale_result = accuracy_score(log_scale_predictions,target_valid)# –∫–∞—á–µ—Å—Ç–≤–æ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–∏
                if log_scale_result > best_log_scale_result:
                    best_log_model_scale = log_model_scale
                    best_log_scale_result = log_scale_result
            except:
                pass
display("Accuracy –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å–∏–∏:", best_log_scale_result)
```

```python
# –°—Ä–∞–∑—É –∂–µ, –≤ —ç—Ç–æ–º —Ä–∞–∑–¥–µ–ª–µ –ø—Ä–æ–≤–µ—Ä–∏–º –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ

test_scale_predictions = best_log_model_scale.predict(feat_test_scale)
display(f'Accuracy –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:{round(accuracy_score(test_scale_predictions,target_test),3)}')
display(f'Precision –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:{round(precision_score(test_scale_predictions, target_test),3)}')
display(f'Recall –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:{round(recall_score(test_scale_predictions, target_test),3)}') 
```

–ù–µ —É–≤–µ—Ä–µ–Ω —á—Ç–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–∏–º–µ–Ω–∏–ª StandartScaler, –Ω–æ –∫–∞–∂–µ—Ç—Å—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–æ—á—Ç–∏ –Ω–µ –∏–∑–º–µ–Ω–∏–ª—Å—è.</div>

–ö–æ–¥ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è –≤–µ—Ä–Ω—ã–π.
</div>

### –ú–µ—Ç–æ–¥ k-–±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π

```python
best_knn_model = None
best_knn_result = 0
      
for alg in ['ball_tree', 'kd_tree', 'brute', 'auto']:
    for metr in ['minkowski', 'euclidean', 'manhattan', 'chebyshev']:
        for neigh in range(1,15):
            knn_model = KNeighborsClassifier(n_neighbors=neigh,
                                            metric=metr,
                                            algorithm=alg)
            knn_model.fit(features_train,target_train) # –æ–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
            knn_predictions = knn_model.predict(features_valid) # –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
            knn_result = accuracy_score(knn_predictions,target_valid)# —Å—á–∏—Ç–∞–µ–º –∫–∞—á–µ—Å—Ç–≤–æ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
            if knn_result > best_knn_result:
                best_knn_model = knn_model
                best_knn_result = knn_result
            
display("Accuracy –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –º–µ—Ç–æ–¥–∞ k-–±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π:", best_knn_result)
```

<div class="alert alert-info">
<b> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞ 2</b>      

–î–æ–±–∞–≤–∏–ª –ø–æ–¥–±–æ—Ä –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –∏ –º–µ—Ç—Ä–∏–∫. –ö–∞–∂–µ—Ç—Å—è accuracy —Å—Ç–∞–ª –ª—É—á—à–µ –Ω–∞ 2% –ø—Ä–∏ –≤–∞–ª–∏–¥–∞—Ü–∏–∏, –ø–æ—Å–º–æ—Ç—Ä–∏–º —á—Ç–æ –±—É–¥–µ—Ç –Ω–∞ —Ç–µ—Å—Ç–µ.</div>

### –ì—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π boosting

```python
best_gbc_model = None
best_gbc_result = 0

for learn in [1, 0.5, 0.25, 0.1, 0.05, 0.01]:
    for est in [1, 2, 4, 8, 16, 32, 64, 100, 200]:
        for depth in range(1,11):
            gbc_model = GradientBoostingClassifier(n_estimators=est, 
                                                    learning_rate=learn,
                                                    max_depth=depth, 
                                                    random_state=12345)
    
            gbc_model.fit(features_train,target_train) # –æ–±—É—á–∞–µ–º –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
            gbc_predictions = gbc_model.predict(features_valid) # –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–∏
            gbc_result = accuracy_score(gbc_predictions,target_valid)# –∫–∞—á–µ—Å—Ç–≤–æ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–∏
            if gbc_result > best_gbc_result:
                best_gbc_model = gbc_model
                best_gbc_result = gbc_result
        
display("Accuracy –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –º–µ—Ç–æ–¥–∞ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ –±—É—Å—Ç–∏–Ω–≥–∞:", best_gbc_result)
```

–ü–æ—Å–ª–µ —Ç–æ–≥–æ, –∫–∞–∫ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–∏ –ø–æ–¥–æ–±—Ä–∞–Ω—ã - —Ç–µ—Å—Ç–∏—Ä—É–µ–º –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö. –ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –Ω–∞ —Ç–µ—Å—Ç–µ (—Å–æ—Ä—Ä–∏ –∑–∞ —Ç–∞–≤—Ç–∞–ª–æ–≥–∏—é) –≤—ã–±–∏—Ä–∞–µ–º –º–æ–¥–µ–ª—å, –∫–æ—Ç–æ—Ä—É—é —Å–º–æ–∂–µ–º –ø–µ—Ä–µ–¥–∞—Ç—å –≤ –ø—Ä–æ–¥–∞–∫—à–Ω. 
</div>

## –ü—Ä–æ–≤–µ—Ä–∫–∞ –º–æ–¥–µ–ª–µ–π –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ

```python
models = [best_tree_model, best_forest_model, best_log_model, best_knn_model, best_gbc_model]

for any_model in models:
    test_predictions = any_model.predict(features_test)
    display(f'–ú–æ–¥–µ–ª—å - {any_model}:')
    display(f'Accuracy –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:{round(accuracy_score(test_predictions,target_test),3)}')
    display(f'Precision –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:{round(precision_score(test_predictions, target_test),3)}')
    display(f'Recall –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:{round(recall_score(test_predictions, target_test),3)}') 
    display('******************************************************************************************************')
```

```python
# –í —Ñ–æ—Ä–º–∞—Ç–µ —Ç–∞–±–ª–∏—Ü—ã

models_name = ['best_tree_model', 'best_forest_model', 'best_log_model', 'best_knn_model', 'best_gbc_model']

common_result = pd.DataFrame({'':['accuracy','precision','recall']})
interm_result = []

for name, any_model in enumerate(models):
    test_predictions = any_model.predict(features_test)
    accuracy = round(accuracy_score(test_predictions,target_test),3)
    precision = round(precision_score(test_predictions, target_test),3)
    recall = round(recall_score(test_predictions, target_test),3) 
    interm_result = pd.Series([accuracy,precision,recall])#interm_result.append(accuracy,precision,recall)
    common_result = common_result.merge(interm_result.rename(models_name[name]), left_index=True, right_index=True)
display(common_result.set_index('').T)
```
